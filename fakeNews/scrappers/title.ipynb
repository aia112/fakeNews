```python
import pandas as pd
from newspaper import Article
import csv
```

```python
df = pd.read_csv("dataset.csv")
df['Authors'].fillna('', inplace = True) ## Replace NaN with blank string
df
```

```python
with open('Titledataset.csv','a',newline='',encoding='utf-8') as new_file:
    csv_writer = csv.writer(new_file)
    csv_writer.writerow(['Authors','Title','Text','URL','Company','Category','Label'])##first uncomment this
```

```python
import newspaper
for i in range(len(df["Authors"])):
    url = df.iloc[i]["URL"]
    try:
     news = Article(url)
     news.download()
     news.parse()
     with open('Titledataset.csv','a',newline='',encoding='utf-8') as new_file:
         csv_writer = csv.writer(new_file)
         csv_writer.writerow([df.iloc[i]["Authors"],news.title,df.iloc[i]["Text"],df.iloc[i]["URL"],df.iloc[i]["Company"],df.iloc[i]["Category"],df.iloc[i]["Label"]])
     print((i/len(df["Authors"]))*100 , "% Complete ", end="\r")
    except:
        pass
print("Done")
```
