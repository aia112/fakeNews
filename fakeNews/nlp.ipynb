```python
from dataClean import clean, removehyphen_underscore, stop
from nltk import word_tokenize
import pandas as pd
import numpy as np
from datetime import datetime
import pickle
```

```python
df = pd.read_csv("dataset/dataset.csv", index_col="Unnamed: 0", encoding="UTF-8")
df
```

```python
list(df["Title"].isnull()).count(True)
```

```python
title = pd.DataFrame({"Title": df["Title"]})
title
```

```python
print(len(stop))
```

```python
tok = list()
for i in range(len(title)):
    tok.append(clean(df.iloc[i]["Title"], absolute = True))
tok
```

```python
title["Tokens"] = pd.Series(tok, index = title.index)
```

```python
title
```

```python
texts = pd.DataFrame({"Text": df["Text"]})
texts
```

```python
tok.clear()
start = datetime.now()
for i in range(len(texts)):
    tok.append(clean(word_tokenize(texts.iloc[i]["Text"]), True))
    print(((i+1)/len(texts))*100, "i: ", i)
tok
print(datetime.now() - start)
```

```python
texts["Tokens"] = pd.Series(tok, index= texts.index)
```

```python
texts
```

```python
def listToString(l1):
    return " ".join(l1)
strings = list()
for i in range(len(title)):
    strings.append(listToString(title.iloc[i]["Tokens"]))
title["TitleString"] = pd.Series(strings, index = title.index)
title
```

```python
strings = list()
for i in range(len(texts)):
    strings.append(listToString(texts.iloc[i]["Tokens"]))
texts["TextString"] = pd.Series(strings, index = texts.index)
texts
```
